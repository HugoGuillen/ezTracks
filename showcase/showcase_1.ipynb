# Code for generating simulated data, configurations, and markdown

````python
import sys
import os
from os import path,listdir
import pandas as pd
import csv
import shutil
import numpy as np
from collections import OrderedDict

sys.path.append('..')
from eztracks import EzConfig

eztracks = '../eztracks.py'
output_data_path = 'simulated_data'
output_img_path = 'simulated_img'
input_gtf = path.join(output_data_path,'simulated_annotation.gtf')
input_gtf2 = path.join(output_data_path,'simulated_annotation_minus.gtf')
dense_tracks_path = path.join(output_data_path,'dense_tracks')
temp_path = 'temp'
ezconfig_ini = ''

for folder in [output_data_path,output_img_path,dense_tracks_path]:
    if not path.exists(folder):
        os.makedirs(folder)
````

## Create simulated data

````python
##############################
# Variables
T_STRAND = '+'  #Strand of transcript
offset = 10**3  #Starting coord for transcript
n_bins = 50     #Number of BED tracks

intron_scale = 10**2
##############################
# Simulated transcript and GTF
exons = [('exon',1,10),
         ('exon',1+3*intron_scale,20+3*intron_scale),
         ('exon',1+(4*intron_scale),70+(4*intron_scale))]
df = pd.DataFrame(exons,columns=['feature','start','end'])
df[['start','end']] = df[['start','end']]+offset
df.loc[len(df.index)] = ['transcript', df['start'].min(), df['end'].max()]
df.sort_values(by='start',inplace=True)

df[['chr','source','score','strand','frame','attr']] = ['chr1','auto','.','+','.','transcript_id "EST01"; transcript_name "transcript01(+)";']
df = df[['chr','source','feature','start','end','score','strand','frame','attr']]
df.to_csv(input_gtf,sep='\t',header=None,index=None,quoting=csv.QUOTE_NONE)

df[['chr','source','score','strand','frame','attr']] = ['chr1','auto','.','-','.','transcript_id "EST01"; transcript_name "transcript01(-)";']
df = df[['chr','source','feature','start','end','score','strand','frame','attr']]
df.to_csv(input_gtf2,sep='\t',header=None,index=None,quoting=csv.QUOTE_NONE)

##############################
# Compute stats
gtf = pd.read_csv(input_gtf,sep='\t',header=None)
gtf = gtf[gtf[2]=='exon']
gtf['L'] = gtf[4]-gtf[3]+1
print('#Region:\t%s:%d-%d(%s)'%(gtf.iloc[0][0],gtf[3].min(),gtf[4].max(),gtf.iloc[0][6]))
print('#Length:\t%d'%(gtf[4].max()-gtf[3].min()))
print('#Exon density:\t%.2e'%(100*gtf['L'].sum()/(gtf[4].max()-gtf[3].min())))

##############################
# Create BED files
t_length = df['end'].max() -  df['start'].min()
coord_start = df['start'].min()-1
coord_end = df['end'].max()
bin_size = int(np.floor(t_length/n_bins))
rows = []
for i in range(n_bins):
    rows.append(('track_%03d'%(i+1),coord_start+i*bin_size,min(coord_start+int((i+1.5)*bin_size),coord_end)))
dfx = pd.DataFrame(rows,columns=['name','start','end'])
dfx['chr'] = 'chr1'
dfx['score'] = 0
dfx['strand'] = ['+' if np.random.rand()<=0.5 else '-' for i in range(len(rows))]
dfx = dfx[['chr','start','end','name','score','strand']]

for name,g in dfx.groupby('name'):    
    output_file = path.join(dense_tracks_path,name+'.bed')
    g.to_csv(output_file,sep='\t',index=None,header=None)    
````

## Create configuration files

````python
params = [
    # 00 Minimal case
    OrderedDict(
        default={'input_gtf':input_gtf,
                 'output_path':temp_path,
                 'region':'chr1:800-1500',},
        plot={},
        template={}),
    
    # 01 Minimal case 2
    OrderedDict(
        default={'input_gtf':input_gtf,
                 'output_path':temp_path,
                 'transcript':'EST01',},
        plot={},
        template={}),
    
    # 02 With tracks
    OrderedDict(
        default={'input_gtf':input_gtf,
                 'output_path':temp_path,
                 'transcript':'EST01',},
        tracks={'Dense_tracks':dense_tracks_path},
        plot={},
        template={}),
    
    # 03 Aesthetics1
    OrderedDict(
        default={'input_gtf':input_gtf,
                 'output_path':temp_path,
                 'transcript':'EST01',},
        tracks={'Dense_tracks':dense_tracks_path},
        plot={'gtf_height':1},
        template={'bedtrack_height':0.2}),
    
    # 04 Aesthetics2
    OrderedDict(
        default={'input_gtf':input_gtf,
                 'output_path':temp_path,
                 'transcript':'EST01',},
        tracks={'Dense_tracks':dense_tracks_path},
        plot={'gtf_height':1,
              'gtf_spacer':0,              
              'color_per_group':False},
        template={'bedtrack_height':0.15,
                  'bedtrack_height_groupname':0.3,
                  'bedtrack_show_name':False,
                  'bedtrack_show_midline':False}),
    
    # 05 Aesthetics3
    OrderedDict(
        default={'input_gtf':input_gtf,
                 'output_path':temp_path,
                 'transcript':'EST01',},
        tracks={'Dense_tracks':dense_tracks_path},
        plot={'gtf_height':1,
              'gtf_spacer':0,              
              'color_per_group':False,
              'colormap_name':'rainbow',
              'colormap_count':50},
        template={'bedtrack_height':0.15,
                  'bedtrack_height_groupname':0.3,
                  'bedtrack_show_name':False,
                  'bedtrack_show_midline':False}),
    
    # 06 Relative
    OrderedDict(
        default={'input_gtf':input_gtf,
                 'output_path':temp_path,
                 'transcript':'EST01',
                 'no_introns':True},
        tracks={'Dense_tracks':dense_tracks_path},
        plot={'gtf_height':1,
              'gtf_spacer':0,              
              'color_per_group':False},
        template={'bedtrack_height':0.5,
                  'bedtrack_height_groupname':0.6,
                  'bedtrack_show_name':True,
                  'bedtrack_show_midline':True}),
    
    # 07 reducing intron space
    OrderedDict(
        default={'input_gtf':input_gtf,
                 'output_path':temp_path,
                 'transcript':'EST01',
                 'no_introns':True,
                 'intron_fraction':0,},
        tracks={'Dense_tracks':dense_tracks_path},
        plot={'gtf_height':1,
              'gtf_spacer':0,              
              'color_per_group':False},
        template={'bedtrack_height':0.5,
                  'bedtrack_height_groupname':0.6,
                  'bedtrack_show_name':True,
                  'bedtrack_show_midline':True}),
    
    # 08 Transcript in negative strand
    OrderedDict(
        default={'input_gtf':input_gtf2,
                 'output_path':temp_path,
                 'transcript':'EST01',},
        tracks={'Dense_tracks':dense_tracks_path},
        plot={'gtf_height':1,
              'gtf_spacer':0,              
              'color_per_group':False},
        template={'bedtrack_height':0.15,
                  'bedtrack_height_groupname':0.3,
                  'bedtrack_show_name':False,
                  'bedtrack_show_midline':False}),
    
    # 09 Intron relative
    OrderedDict(
        default={'input_gtf':input_gtf2,
                 'output_path':temp_path,
                 'transcript':'EST01',
                 'no_introns':True,
                 'intron_fraction':0,},
        tracks={'Dense_tracks':dense_tracks_path},
        plot={'gtf_height':1,
              'gtf_spacer':0,              
              'color_per_group':False},
        template={'bedtrack_height':0.5,
                  'bedtrack_height_groupname':0.6,
                  'bedtrack_show_name':True,
                  'bedtrack_show_midline':True}),
]

print('activate eztracks')
print('cd',path.abspath('.'),'\n')
for idx,param in enumerate(params):    
    config_file = path.join(output_data_path,'ezconfig_%02d.ini'%idx)
    output_img = 'img_%02d.png'%idx
    param['plot']['output_img'] = output_img
    EzConfig.write_config(config_file,param)
    print('python3',eztracks,'prepare',config_file)
    print('python3',eztracks,'draw',config_file)
    print('cp',path.join(temp_path,output_img),path.join(output_img_path,output_img))
    print('rm -r',temp_path,'\n')
````

````python
for idx in range(len(params)):
    config_file = path.join(output_data_path,'ezconfig_%02d.ini'%idx)
    with open(config_file,'r') as f:
        lines = f.readlines()
    print('## Example',idx+1)
    print('\ntext\n')
    print('[configuration file](%s):'%(config_file))
    print('```ini')    
    print(''.join(lines))
    print('```\n')
    print('**commands:**')
    print('```bash\npython ../eztracks.py prepare %s\npython ../eztracks.py draw %s\n```\n'%(config_file,config_file))
    print("![](%s)"%path.join(output_img_path,'img_%02d.png'%idx))    
````
